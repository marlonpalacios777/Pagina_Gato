```python
import matplotlib.pyplot as plt

# Datos de la simulación para 6 sprints
sprints = [1, 2, 3, 4, 5, 6]

# Tasa de éxito de envío (en porcentaje)
tasa_exito_envio = [70, 75, 80, 85, 90, 95]

# Número de errores de validación
errores_validacion = [15, 12, 10, 8, 5, 3]

# Verificación de Layout y Diseño (en porcentaje de aprobación)
verificacion_layout = [60, 65, 70, 75, 80, 85]

# Verificación de Funcionalidad (en porcentaje de aprobación)
verificacion_funcionalidad = [50, 60, 70, 80, 90, 95]

# Verificación de Compatibilidad (en porcentaje de aprobación)
verificacion_compatibilidad = [40, 50, 60, 75, 85, 90]

# Documentación de Resultados (en porcentaje de completitud)
documentacion_resultados = [30, 50, 70, 80, 90, 100]

# Satisfacción del Usuario (escala de 1 a 10)
satisfaccion_usuario = [4, 5, 6, 7, 8, 9]

# Creación de las gráficas
fig, axs = plt.subplots(4, 2, figsize=(14, 14))

# Gráfica de Tasa de Éxito de Envío
axs[0, 0].plot(sprints, tasa_exito_envio, marker='o')
axs[0, 0].set_title('Tasa de Éxito de Envío')
axs[0, 0].set_xlabel('Sprints')
axs[0, 0].set_ylabel('Tasa de Éxito (%)')

# Gráfica de Número de Errores de Validación
axs[0, 1].plot(sprints, errores_validacion, marker='o', color='r')
axs[0, 1].set_title('Número de Errores de Validación')
axs[0, 1].set_xlabel('Sprints')
axs[0, 1].set_ylabel('Errores de Validación')

# Gráfica de Verificación de Layout y Diseño
axs[1, 0].plot(sprints, verificacion_layout, marker='o', color='g')
axs[1, 0].set_title('Verificación de Layout y Diseño')
axs[1, 0].set_xlabel('Sprints')
axs[1, 0].set_ylabel('Aprobación Layout (%)')

# Gráfica de Verificación de Funcionalidad
axs[1, 1].plot(sprints, verificacion_funcionalidad, marker='o', color='m')
axs[1, 1].set_title('Verificación de Funcionalidad')
axs[1, 1].set_xlabel('Sprints')
axs[1, 1].set_ylabel('Aprobación Funcionalidad (%)')

# Gráfica de Verificación de Compatibilidad
axs[2, 0].plot(sprints, verificacion_compatibilidad, marker='o', color='c')
axs[2, 0].set_title('Verificación de Compatibilidad')
axs[2, 0].set_xlabel('Sprints')
axs[2, 0].set_ylabel('Aprobación Compatibilidad (%)')

# Gráfica de Documentación de Resultados
axs[2, 1].plot(sprints, documentacion_resultados, marker='o', color='b')
axs[2, 1].set_title('Documentación de Resultados')
axs[2, 1].set_xlabel('Sprints')
axs[2, 1].set_ylabel('Completitud Documentación (%)')

# Gráfica de Satisfacción del Usuario
axs[3, 0].plot(sprints, satisfaccion_usuario, marker='o', color='y')
axs[3, 0].set_title('Satisfacción del Usuario')
axs[3, 0].set_xlabel('Sprints')
axs[3, 0].set_ylabel('Satisfacción (1-10)')

# Ajustar layout para una mejor visualización
fig.tight_layout()
plt.show()
```
